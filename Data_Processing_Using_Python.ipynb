```python
# 1. Create a bar chart to show the distribution of passengers by class (1st, 2nd, 3rd)

class_counts = df['Pclass'].value_counts()
class_counts.plot(kind='bar', title='Passenger Distribution by Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# 3. Calculate and visualize the survival rate of passengers with a pie chart

survival_counts = df['Survived'].value_counts()
plt.pie(survival_counts, labels=['Did not Survive', 'Survived'], autopct='%1.1f%%', startangle=90)
plt.title('Survival Rate of Passengers')
plt.axis('equal')
plt.show()
```

```python
# 4. Create a bar chart to show the count of passengers by gender

gender_counts = df['Sex'].value_counts()
gender_counts.plot(kind='bar', title='Passenger Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# 5. Create a stacked bar chart to visualize the survival rate by gender

survival_gender = df.groupby(['Sex', 'Survived'])['PassengerId'].count().unstack()
survival_gender.plot(kind='bar', stacked=True, title='Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.legend(['Did not Survive', 'Survived'])
plt.show()
```

```python
# 6. Create a count plot (bar chart) to show the number of passengers who embarked from each location (S, C, Q)

sns.countplot(data=df, x='Embarked', order=['S', 'C', 'Q'])
plt.title('Passengers Embarked from Each Location')
plt.xlabel('Embarkation Point')
plt.ylabel('Count')
plt.show()
```

```python
# 8. Create a scatter plot to visualize the relationship between fare and age

plt.scatter(df['Age'], df['Fare'], alpha=0.5)
plt.title('Relationship Between Fare and Age')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

# **Part 3 - Cleaning**

```python
# Load the dataset

df = pd.read_csv("Corona_NLP.csv")

# 1. Remove hyperlinks
df['CleanedTweet'] = df['OriginalTweet'].str.replace(r'http\S+', '', regex=True)

# 2. Remove hashtags
df['CleanedTweet'] = df['CleanedTweet'].str.replace(r'#\w+', '', regex=True)

# 3. Replace \n and \r with an empty space
df['CleanedTweet'] = df['CleanedTweet'].str.replace('\n', ' ').str.replace('\r', ' ')

# Fill NaN values with an empty string
df['CleanedTweet'].fillna('', inplace=True)

# 4. Tokenization (basic splitting by whitespace)
df['CleanedTweet'] = df['CleanedTweet'].str.split()

# 5. Remove stopwords and punctuation signs
stopwords = set(['the', 'and', 'in', 'to', 'of'])
df['CleanedTweet'] = df['CleanedTweet'].apply(lambda tokens: [word for word in tokens if isinstance(word, str) and word.lower() not in stopwords and word.isalpha()])

# 6. Perform simple stemming (first 4 characters of each word)
df['CleanedTweet'] = df['CleanedTweet'].apply(lambda tokens: [word[:4] for word in tokens])

# Convert the tokenized words back to a string
df['CleanedTweet'] = df['CleanedTweet'].apply(lambda tokens: ' '.join(tokens))

# Visualize the first 5 rows of the cleaned dataset
df['CleanedTweet'].head()
```

# 3. Fit a regression line and visualise the data points.

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
# Load the data
data = pd.read_csv('q3_data.csv')
x1 = data['x1']
x2 = data['x2']

# Create and fit a linear regression model
reg = LinearRegression().fit(x1.values.reshape(-1, 1), x2)

# Make predictions using the model
x1_pred = np.array([min(x1), max(x1)]).reshape(-1, 1)
x2_pred = reg.predict(x1_pred)

# Plot the data points and regression line
plt.scatter(x1, x2, label='Data Points')
plt.plot(x1_pred, x2_pred, color='red', label='Regression Line')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend()
plt.title('Scatter Plot and Regression Line')
plt.show()
```
